# fillna()
# he fillna() function in Python is part of the pandas library and is used to fill missing (NaN) values in a DataFrame or Series with a specified value, method, or strategy. This is particularly useful for cleaning and preparing data for analysis, where missing values can impact calculations and results.

Syntax
python
Copy code
DataFrame.fillna(value=None, method=None, axis=None, inplace=False, limit=None)
Parameters
value: The value to replace missing values with. It can be a scalar, dictionary, or Series.
method: The method to fill values ('ffill' for forward fill, 'bfill' for backward fill).
axis: Determines if filling is done across rows (0) or columns (1).
inplace: If True, modifies the original DataFrame. If False, returns a new DataFrame with filled values.
limit: Maximum number of NaNs to fill.
Examples
Filling Missing Values with a Constant

python
Copy code
import pandas as pd
import numpy as np

df = pd.DataFrame({'A': [1, np.nan, 3], 'B': [4, 5, np.nan]})
df_filled = df.fillna(0)
print(df_filled)
# Output:
#      A    B
# 0  1.0  4.0
# 1  0.0  5.0
# 2  3.0  0.0
Forward Fill (Propagate Last Valid Value)

python
Copy code
df_filled_ffill = df.fillna(method='ffill')
print(df_filled_ffill)
# Output:
#      A    B
# 0  1.0  4.0
# 1  1.0  5.0
# 2  3.0  5.0
Backward Fill (Propagate Next Valid Value)

python
Copy code
df_filled_bfill = df.fillna(method='bfill')
print(df_filled_bfill)
# Output:
#      A    B
# 0  1.0  4.0
# 1  3.0  5.0
# 2  3.0  NaN
Filling with Different Values for Each Column

python
Copy code
df_filled_diff = df.fillna({'A': 0, 'B': 10})

# 1  0.0   5.0
# 2  3.0  10.0

# Types of data analysis

Univariate, bivariate, and multivariate analyses are fundamental statistical techniques used to explore and analyze data based on the number of variables involved:

1. Univariate Analysis
Purpose: Analyzes a single variable to understand its distribution, central tendency, and spread.
Techniques:
Descriptive Statistics: Calculates measures like mean, median, mode, range, variance, and standard deviation.
Frequency Distribution: Summarizes the data into categories to observe distribution.
Visualization:
Histograms: Shows frequency distribution.
Box Plots: Visualizes the spread and identifies outliers.
Pie Charts: Shows proportions for categorical data.
Example: Analyzing the ages of students in a class.
2. Bivariate Analysis
Purpose: Examines the relationship between two variables to determine if they are related or influence one another.
Techniques and Subtypes:
Correlation Analysis: Measures the strength and direction of the relationship between two variables.
Pearson Correlation: Measures linear relationships for continuous variables.
Spearman’s Rank Correlation: Measures monotonic relationships, often used for ordinal or non-normally distributed data.
Regression Analysis: Explores the dependency of one variable on another.
Simple Linear Regression: Studies the linear relationship between two continuous variables.
Logistic Regression: Used when the dependent variable is binary.
Visualization:
Scatter Plots: Visualizes the relationship and trends between two variables.
Line Charts: Useful for time series analysis when examining changes over time.
Example: Exploring the relationship between study hours and exam scores.
3. Multivariate Analysis
Purpose: Analyzes more than two variables to explore complex relationships or effects in a dataset.
Techniques and Subtypes:
Multiple Regression: Extends simple linear regression to study the effect of multiple independent variables on a single dependent variable.
Factor Analysis: Reduces data dimensionality by grouping related variables into factors.
Principal Component Analysis (PCA): Reduces the number of variables by identifying principal components that capture the most variance.
Exploratory Factor Analysis (EFA): Identifies underlying relationships between variables and groups them into factors.
Cluster Analysis: Groups similar data points based on selected features.
K-Means Clustering: Partitions data into clusters based on centroids.
Hierarchical Clustering: Builds a hierarchy of clusters through a dendrogram.
MANOVA (Multivariate Analysis of Variance): Extends ANOVA to compare means across groups with multiple dependent variables.
Discriminant Analysis: Classifies observations into predefined categories based on predictor variables.
Visualization:
3D Scatter Plots: Useful for visualizing relationships among three continuous variables.
Heatmaps: Displays correlation matrices to show relationships between multiple variables.
Example: Analyzing how income, age, and education affect purchasing behavior in a survey dataset.
Each of these techniques provides unique insights into data and is chosen based on the complexity of relationships between variables and the specific questions being asked in the analysis.


There are several types of data analysis techniques, each suited for different purposes and types of data. Here’s an overview of the main types:

1. Descriptive Analysis
Purpose: Summarizes historical data to understand what has happened.
Methods: Uses measures like mean, median, mode, and standard deviation; also includes visualization techniques like charts, tables, and graphs.
Example: Reporting sales figures over the last quarter to identify patterns or trends.
2. Diagnostic Analysis
Purpose: Explores historical data to determine the reasons behind certain outcomes.
Methods: Often uses drill-down methods, correlation analysis, and data mining techniques.
Example: Analyzing a drop in sales by looking at different factors like price changes, customer demographics, or seasonality.
3. Predictive Analysis
Purpose: Uses historical data to predict future outcomes or trends.
Methods: Utilizes statistical modeling, machine learning algorithms, and forecasting techniques.
Example: Forecasting future sales based on past sales data and market conditions.
4. Prescriptive Analysis
Purpose: Recommends actions to achieve desired outcomes.
Methods: Involves optimization algorithms, simulation, and machine learning models that suggest solutions based on predictive analysis results.
Example: A recommendation system that suggests the optimal inventory levels based on predicted demand.
5. Exploratory Analysis
Purpose: Investigates data without predetermined assumptions, usually to find hidden patterns or relationships.
Methods: Uses visualization tools, clustering, and factor analysis.
Example: Discovering customer segments based on purchasing behavior.
6. Inferential Analysis
Purpose: Makes inferences about a larger population based on a sample of data.
Methods: Relies on hypothesis testing, confidence intervals, and other statistical tests.
Example: Analyzing survey data to infer general opinions about a product among a larger population.
7. Causal Analysis
Purpose: Determines if one event causes another.
Methods: Uses methods like A/B testing, controlled experiments, and causal inference techniques.
Example: Determining if a marketing campaign is the cause of increased sales.
8. Mechanistic Analysis
Purpose: Understands the exact changes in variables that lead to changes in other variables, often aiming to establish a deterministic relationship.
Methods: Used in scientific research and engineering, with a focus on exact outcomes.
Example: In engineering, analyzing how temperature affects material strength.
9. Text Analysis (Text Mining)
Purpose: Extracts information from text data and converts it into structured data.
Methods: Involves natural language processing (NLP), sentiment analysis, and keyword extraction.
Example: Analyzing customer feedback to identify common complaints.
10. Network Analysis
Purpose: Examines relationships between interconnected entities.
Methods: Uses graph theory and network visualization.
Example: Analyzing social network data to find key influencers within a network.
11. Time Series Analysis
Purpose: Analyzes data points collected or recorded at specific time intervals to identify trends, seasonal patterns, or cycles.
Methods: Includes decomposition, moving averages, ARIMA modeling, etc.
Example: Forecasting stock prices or energy demand over time.
Each type of analysis is chosen based on the objective of the study and the nature of the data, helping organizations make informed decisions based on both historical and current data patterns.

# Generator Function

# The yield a statement produces the current Fibonacci number and then suspends the function until the next call.

The yield statement in Python is used in the context of a generator function. It allows the function to produce a series of values over multiple calls, maintaining its state between calls.
When a generator function is called, it doesn't execute the function immediately. Instead, it returns an iterator, and the function is executed only when the iterator's next() method is called.

```python
def fibonacci_generator():
    a, b = 0, 1
    while True:
        yield a
        a, b = b, a + b

# Example usage:
fibonacci = fibonacci_generator()

# Print the first 5 Fibonacci numbers
for i in range(5):
    print(next(fibonacci))
```

```python
def fibo():
    n1,n2=0,1
    while True:
        yield n1
        n1,n2=n2,n1+n2

f=fibo()
for i in range(10):
    print(next(f))
```

# Inheritance in Class

```python
# Single inheritance
class person:
    def __init__(self,name,email,contact):
        self.name=name
        self.email=email
        self.contact=contact
    def show(self):
        print("Name=",self.name)
        print("Email=",self.email)
        print("Email=",self.contact)
```

```python
p1=person("Hiten","hiten@gmail.com",9898989898)
p1.show()
```

```python
# Create parent class, use super().__init__(name,email,contact) (self removed)
class Employee(person):
    def __init__(self,name,email,contact,company,designation,salary):
        super().__init__(name,email,contact)
        self.company=company
        self.designation=designation
        self.salary=salary
    def display(self):
        super().show()
        print("Company=",self.company)
        print("Designation=",self.designation)
        print("Salary=",self.salary)
```

```python
"""In inheritance we create object of child/sub class that is the reason, you 
have to invoke super/parent class constructor in child class"""
```

```python
e1=Employee("Gargi","gargi@gmail.com",9890989890,"TCS","Developer",28000)
e1.display()
```

```python
# Create Multilevel/ grand parents inheritance, use  sequence in first show(self). second display(self): super().show(). third showdetails(self): super().display() 
class Manager(Employee):
    def __init__(self,name,email,contact,company,designation,salary,department):
        super().__init__(name,email,contact,company,designation,salary)
        self.department=department
    def showdetails(self):
        super().display()
        print("Department=",self.department)
```

```python
m1=Manager("Diksha","diksha@gmai.com",9878767898,"ETLHIVE","BranchManager",65000,"Sales")
m1.showdetails()
```

# Heirarchical

```python
class vehicle:
    def __init__(self,manufacturer,model):
        self.manufacturer=manufacturer
        self.model=model
    def showvehicle(self):
        print(f"Vehicle is manufactured by{self.manufacturer},Model is {self.model}")
```

```python
class twowheeler(vehicle):
    def __init__(self,manufacturer,model,price):
        super().__init__(manufacturer,model)
        self.price=price
    def showbike(self):
        super().showvehicle()
        print("Price=",self.price)
```

```python
class fourwheeler(vehicle):
    def __init__(self,manufacturer,model,price,AirBags):
        super().__init__(manufacturer,model)
        self.price=price
        self.AirBags=AirBags
    def showcar(self):
        super().showvehicle()
        print("Price=",self.price)
        print(f"Has {self.AirBags} AirBags")
```

```python
b=twowheeler("Honda","Aviator",120000)
i10=fourwheeler("Honda","i10",700000,"Driver and Passenger")
```

```python
b.showbike()
```

```python
i10.showcar()
```
